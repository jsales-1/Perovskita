```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("perovskites_parte1/arquivo_1_1000.txt",sep='\t')
stats = np.arange(1001,78002,1000)
ends = np.arange(2000,80000,1000)
ends[-1] = 78557
for i,j in zip(stats,ends):
    print(i,end='\r')
    df_add = pd.read_csv(f"perovskites_parte1/arquivo_{i}_{j}.txt",sep='\t')
    df1 = pd.concat([df1, df_add], ignore_index=True)
```

```python
df2 = pd.read_csv("perovskites_parte2/arquivo_1_1000.txt",sep='\t')
stats = np.arange(1001,80002,1000)
ends = np.arange(2000,81001,1000)
ends[-1] = 80828
for i,j in zip(stats,ends):
    print(i,end='\r')
    df_add = pd.read_csv(f"perovskites_parte2/arquivo_{i}_{j}.txt", sep='\t', on_bad_lines='skip')
    df2 = pd.concat([df2, df_add], ignore_index=True)
```
